# 📅 DAY 1

## 1️⃣ Python Basics (10 min)

```python
# Variables
name = "Ali"
age = 20
score = 85.5
print(name, age, score)
```

```python
# Lists + Loop
scores = [70, 85, 90, 60, 95]

for s in scores:
    print(s, "-> Pass" if s >= 70 else "-> Fail")
```

```python
# Function
def average(nums):
    return sum(nums) / len(nums)

print("Average score:", average(scores))
```

## 2️⃣ NumPy Basics (30 min)

```python
import numpy as np

# Create an array
arr = np.array([10, 20, 30, 40, 50])
print("Array:", arr)
print("Shape:", arr.shape)
print("Type:", arr.dtype)
```

```python
# Useful array creators
print(np.zeros(5))
print(np.ones(5))
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5))
```

```python
# Indexing & Slicing
arr = np.array([5, 15, 25, 35, 45])
print("First:", arr[0])
print("Last:", arr[-1])
print("First 3:", arr[:3])
print("Greater than 20:", arr[arr > 20])
```

```python
# Math operations
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print("a + b =", a + b)
print("a * 2 =", a * 2)
print("sum   =", a.sum())
print("mean  =", a.mean())
```

```python
# Statistics on random data
np.random.seed(1)
data = np.random.randint(50, 100, 10)

print("Data:", data)
print("Mean:", data.mean())
print("Max :", data.max())
print("Min :", data.min())
```

## 3️⃣ Matplotlib Basics (40 min)

```python
import matplotlib.pyplot as plt

# Line chart
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o', color='blue')
plt.title("Line Chart")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
# Bar chart
subjects = ['Math', 'Science', 'English']
marks = [80, 90, 70]

plt.bar(subjects, marks, color=['blue', 'green', 'orange'])
plt.title("Marks by Subject")
plt.ylabel("Marks")
plt.show()
```

```python
# Histogram
data = np.random.normal(50, 10, 200)

plt.hist(data, bins=20, color='purple')
plt.title("Histogram")
plt.show()
```

```python
# Scatter plot
x = np.random.rand(30)
y = np.random.rand(30)

plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.show()
```

```python
# Pie chart
labels = ['Mobile', 'Desktop', 'Tablet']
values = [50, 35, 15]

plt.pie(values, labels=labels, autopct='%1.0f%%')
plt.title("Device Usage")
plt.show()

print("✅ Day 1 done!")
```

# 📅 DAY 2

## 1️⃣ Load a Dataset (15 min)

```python
import pandas as pd

# Load a free built-in dataset (no download needed)
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

print("Rows, Columns:", df.shape)
df.head()
```

## 2️⃣ Explore with Pandas (30 min)

```python
# Basic info
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Average tip by day
df.groupby('day')['tip'].mean()
```

```python
# Filter: bills above $20
big_bills = df[df['total_bill'] > 20]
print("Number of big bills:", len(big_bills))
big_bills.head()
```

## 3️⃣ Visualize the Data (25 min)

```python
# Bar chart: average tip per day
avg_tip = df.groupby('day')['tip'].mean()

plt.bar(avg_tip.index, avg_tip.values, color='teal')
plt.title("Average Tip by Day")
plt.ylabel("Tip ($)")
plt.show()
```

```python
# Scatter plot: total bill vs tip
plt.scatter(df['total_bill'], df['tip'], color='darkorange', alpha=0.6)
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
# Histogram: distribution of total bill
plt.hist(df['total_bill'], bins=20, color='seagreen')
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

## 4️⃣ Mini Dashboard (40 min)

```python
# A simple 4-chart dashboard in one figure
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Restaurant Tips Dashboard", fontsize=16, fontweight='bold')

# 1. Average tip by day
avg_tip = df.groupby('day')['tip'].mean()
axes[0, 0].bar(avg_tip.index, avg_tip.values, color='teal')
axes[0, 0].set_title("Average Tip by Day")

# 2. Total bill distribution
axes[0, 1].hist(df['total_bill'], bins=15, color='seagreen')
axes[0, 1].set_title("Total Bill Distribution")

# 3. Total bill vs tip
axes[1, 0].scatter(df['total_bill'], df['tip'], color='darkorange', alpha=0.6)
axes[1, 0].set_title("Bill vs Tip")

# 4. Number of customers by day
counts = df['day'].value_counts()
axes[1, 1].pie(counts, labels=counts.index, autopct='%1.0f%%')
axes[1, 1].set_title("Customers by Day")

plt.tight_layout()
plt.savefig("dashboard.png")
plt.show()

print("✅ Dashboard saved as dashboard.png")
print("🎉 Day 2 done! Bootcamp complete!")
```
