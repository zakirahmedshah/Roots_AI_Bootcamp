## 🧰 Step 1: Get Our Tools Ready

Before we cook, we need our kitchen tools 🍳. In Python, our "tools" are called **libraries**. Here's what each one does:

| Library | What it does | Why we need it |
|---|---|---|
| **NumPy** | Works with numbers & random data | To create fake exam scores, study hours, etc. |
| **Pandas** | Organizes data into tables | To store all student info neatly (like Excel) |
| **Matplotlib** | Draws charts & graphs | To turn numbers into pictures! |

```python
import numpy as np              # for numbers & random data
import pandas as pd             # for tables (DataFrames)
import matplotlib.pyplot as plt # for charts

plt.style.use('seaborn-v0_8-darkgrid')  # makes charts look nicer
print("✅ Tools loaded! Let's go 🚀")
```

---
## 🎲 Step 2: Create Our Student Data

We'll imagine **100 students** and randomly generate:

- ⏰ How many hours they study per day
- 📝 Their exam score
- 🎮 Their favorite game
- 🎨 Their hobby

`np.random` lets us generate realistic random numbers — `np.random.seed(42)` just makes sure everyone gets the **same random data**, so our results match.

```python
np.random.seed(42)
n = 100  # number of students

# How many hours each student studies (1 to 8 hours)
study_hours = np.random.uniform(1, 8, n).round(1)

# Exam score: more study = higher score (with some randomness)
exam_score = (study_hours * 9 + np.random.normal(20, 10, n)).clip(30, 100).round(1)

# Favorite games & hobbies (chosen randomly from a list)
favorite_game = np.random.choice(['PUBG', 'Free Fire', 'Minecraft', 'FIFA', 'Chess'], n)
hobby = np.random.choice(['Gaming', 'Reading', 'Sports', 'Music'], n)

# Put it all into a table (DataFrame)
df = pd.DataFrame({
    'StudyHours': study_hours,
    'ExamScore': exam_score,
    'FavoriteGame': favorite_game,
    'Hobby': hobby
})

print(f"✅ Created data for {n} students!")
df.head()
```

---
## 🔍 Step 3: Quick Surprise Facts

Before drawing charts, let's ask Pandas a few quick questions about our data — you might be surprised by the answers! 😲

```python
print("📊 Average exam score :", df['ExamScore'].mean().round(1))
print("🎮 Most popular game  :", df['FavoriteGame'].mode()[0])
print("🎨 Most popular hobby :", df['Hobby'].mode()[0])
print("🏆 Highest score      :", df['ExamScore'].max())
```

---
## 📊 Step 4: Let's Make Charts!

### Chart 1 — 🎮 What games do students like most?

We use `plt.bar()` to make a **bar chart** — great for comparing categories (like games).

```python
game_counts = df['FavoriteGame'].value_counts()  # count how many students like each game

plt.figure(figsize=(7, 4))
plt.bar(game_counts.index, game_counts.values, color='skyblue')
plt.title('🎮 Favorite Games')
plt.ylabel('Number of Students')
plt.show()
```

### Chart 2 — 🎨 What do students do for fun?

`plt.pie()` makes a **pie chart** — perfect for showing **percentages** of a whole.

```python
hobby_counts = df['Hobby'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(hobby_counts.values, labels=hobby_counts.index, autopct='%1.0f%%', startangle=90)
plt.title('🎨 Student Hobbies')
plt.show()
```

### Chart 3 — 📝 How did everyone score?

`plt.hist()` makes a **histogram** — it shows how values are spread out (e.g. most students score around 60-80).

```python
plt.figure(figsize=(7, 4))
plt.hist(df['ExamScore'], bins=15, color='mediumseagreen', edgecolor='white')
plt.title('📝 Exam Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

### Chart 4 — 🤔 The BIG question: Does studying more = better scores?

`plt.scatter()` makes a **scatter plot** — perfect for seeing if **two things are related**.

```python
plt.figure(figsize=(7, 4))
plt.scatter(df['StudyHours'], df['ExamScore'], color='orange', alpha=0.7)
plt.title('⏰ Study Hours vs Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()

print("👀 Look closely... do you see a pattern? More study hours usually = higher scores!")
```

---
## 🏆 Step 5: Combine Everything into ONE Dashboard

Now for the fun part — let's put **all 4 charts together** in a single dashboard using `plt.subplots()`. This creates a **grid of charts** (2 rows × 2 columns).

```python
fig, axes = plt.subplots(2, 2, figsize=(11, 9))
fig.suptitle('🌱 ROOTS AI Bootcamp — Student Life Dashboard', fontsize=16, fontweight='bold')

# Chart 1: Favorite games
axes[0, 0].bar(game_counts.index, game_counts.values, color='skyblue')
axes[0, 0].set_title('🎮 Favorite Games')
axes[0, 0].tick_params(axis='x', rotation=30)

# Chart 2: Hobbies
axes[0, 1].pie(hobby_counts.values, labels=hobby_counts.index, autopct='%1.0f%%')
axes[0, 1].set_title('🎨 Hobbies')

# Chart 3: Exam scores
axes[1, 0].hist(df['ExamScore'], bins=15, color='mediumseagreen', edgecolor='white')
axes[1, 0].set_title('📝 Exam Scores')

# Chart 4: Study hours vs score
axes[1, 1].scatter(df['StudyHours'], df['ExamScore'], color='orange', alpha=0.7)
axes[1, 1].set_title('⏰ Study Hours vs Score')

plt.tight_layout()
plt.savefig('roots_dashboard.png')
plt.show()

print("🎉 Congrats! You just built Project 1 of ROOTS AI Bootcamp!")
```
